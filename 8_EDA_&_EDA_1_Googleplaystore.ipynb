```python
# Interview Questions: 

# Summarisation of Plots: 

# 1. Scatter Plot: 
# Where do we use scatter plot? 
# Use when you want to visualize the relationship between two continuous variables.
# Useful for identifying patterns, correlations, or trends between variables.
# Each point represents an observation in the dataset, making it suitable for displaying individual data points.
    
# 2. Line chart: 
# Where do we use line chart? 
# Whenever on x axis we have time component 
# Example: Stock Price 
# Use when you want to visualize trends or patterns over time or across ordered categories.
# Useful for showing how a variable changes over a continuous or ordered variable (e.g., time).
# Often used for time series data or data with a natural ordering.

# 3. bar chart: 
# Where do we use bar chart? 
# Use when you have categorical data and want to compare the frequency or count of different categories.
# Useful for showing comparisons between discrete categories or groups.
# Can be horizontal or vertical depending on the number of categories and the length of category labels.

# 4. count chart: 
# Where do we use count chart? 
# It is used for categorical data 
# It is a type of bar plot that shows the frequency or count of observations in each category of a categorical variable.

# 5. pie chart: 
# Where do we use pie chart? 
# It is used for categorical data 
# distribution of different unique classes 
# Use sparingly when you want to visualize the proportion of different categories within a whole.
# Useful for showing relative sizes or proportions of categories.
# Best suited for a small number of categories (preferably less than 5 or 6) to avoid clutter.

# 6. box plot and violin plot: Advance plot 
# where do we use box plot and violin plot? 
# Comparison of Distributions: 
#   When you need to compare the distribution of a continuous variable across different categories or groups.
# Comparing Multiple Groups: 
#   When you have multiple groups or categories to compare, 
#   box plots can effectively display the distribution of the continuous variable within each group side by side, 
#   allowing for easy comparison.
# To determine for getting outlier: extreme values 
# Use when you want to visualize the distribution, spread, and variability of a dataset.
# Useful for identifying outliers, comparing distributions, and understanding the central tendency of data.
# Provides information about median, quartiles, and potential outliers.

# violin plot:
# Similar to a box plot but also includes a kernel density estimation (KDE) to show the probability density of the data at different values.
# Useful for visualizing the distribution of data, especially when comparing multiple groups or categories.
# Provides more information about the shape of the distribution compared to a box plot.

# 7. histogram/distribution/kde plot:  
# where do we use histogram/distribution plot? 
# continuous of data 
# Use when you have numerical data and want to visualize the distribution of that data.
# Useful for understanding the shape, center, and spread of a dataset.
# Typically used for continuous data where the values are grouped into bins.

# 8. heat map: 
# where do we use heat map? 
# understand corelation like how one feature is behaving with another feature 
# Use when you want to visualize the relationship between two categorical variables or two numerical variables.
# Useful for identifying patterns or correlations in large datasets.
# Colors represent the intensity or value of a third variable, making it easy to identify patterns.

# 9. pair plot: 
# where do we use? 
# all the numerical relationship in one plot 
# It is a grid of scatterplots where each scatterplot shows the relationship between pairs of variables in a dataset, 
#    useful for exploring correlations and distributions.

# 10. join plot: 
# where do we use? 
# relationship and distribution of data 
# similar to scatter plot 
# It gives distribution of data addition to scatter plot 
# It combines scatterplot and histograms to visualize the relationship between two variables and their individual distributions, 
#    helpful for understanding bivariate relationships.

# 11. swarmplot/relplot: 
# where do we use? 
# similar to scatter plot 
# It displays categorical data with points spread out along the categorical axis, 
#    providing a clearer view of the distribution of values compared to traditional scatter plots.

# 12. LMplot: 
# where do we use? 
# similar to regression line 
# LMPlot is a seaborn function that creates a scatterplot with a linear regression line fit to the data, 
# useful for visualizing the relationship between two variables and assessing the strength and direction of the relationship.

# 13. Stacked bar chart 
# cross tab : when two columns having same values like yes, no 
```
