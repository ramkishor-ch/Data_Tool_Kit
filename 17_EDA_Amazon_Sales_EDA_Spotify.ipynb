```python
# 1. What is the average rating for each product category?

import matplotlib.pyplot as plt

df_cp.columns
# Group by 'product_category' and calculate the mean of 'rating'
average_ratings = df_cp.groupby('category')['rating'].mean()[:20]
# average_ratings

# Plot the average ratings
plt.figure(figsize=(10, 6))
average_ratings.plot(kind='bar', color='skyblue')
plt.title('Average Rating by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 2. What are the top rating_count products by category?

# df_cp.groupby('category')['rating_count'].max(ascending=False)
# average_ratings = df.groupby('product_category')['rating'].mean()

# Sort the DataFrame by 'rating_count' within each category group
sorted_df = df_cp.sort_values(by='rating_count', ascending=False)

# Group by 'product_category' and select the top product in each category
top_products_by_category = sorted_df.groupby('category').head()

# Create a bar plot
plt.figure(figsize=(10, 6))
top_products_by_category.plot(kind='bar', x='category', y='rating_count', color='skyblue')
plt.title('Top Rated Products by Category')
plt.xlabel('Product Category')
plt.ylabel('Rating Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 3. What is the distribution of discounted prices vs. actual prices?
import seaborn as sns

# Assuming you have two columns: 'actual_price' and 'discounted_price' in your dataset

# Plotting a histogram for actual prices
plt.figure(figsize=(12, 9))
sns.histplot(data=df_cp, x='actual_price_rupees', color='skyblue', kde=True, label='Actual Prices')

# Plotting a histogram for discounted prices
sns.histplot(data=df_cp, x='discounted_price_rupees', color='orange', kde=True, label='Discounted Prices')

# Adding labels and title
plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Displaying the plot
plt.legend()
plt.show()
```

```python
# 4. How does the average discount percentage vary across categories?

# Assuming you have a column 'discount_percentage' and 'category' in your dataset

# Calculate the average discount percentage for each category
average_discount_by_category = df_cp.groupby('category')['discount_percentage'].mean().reset_index()

# Plotting the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(data=average_discount_by_category, x='category', y='discount_percentage', palette='viridis')

# Adding labels and title
plt.title('Average Discount Percentage Across Categories')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
# 5. What are the most popular product names?

# Assuming you have a column 'product_name' in your dataset

# Count the occurrences of each product name
product_name_counts = df_cp['product_name'].value_counts().reset_index()

# Rename the columns for better readability
product_name_counts.columns = ['Product Name', 'Count']

# Sort the dataframe by count in descending order to get the most popular products first
product_name_counts = product_name_counts.sort_values(by='Count', ascending=False)

# Plotting the horizontal bar plot
plt.figure(figsize=(10, 8))
sns.barplot(data=product_name_counts.head(10), x='Count', y='Product Name', palette='magma')

# Adding labels and title
plt.title('Top 10 Most Popular Product Names')
plt.xlabel('Count')
plt.ylabel('Product Name')

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
# 6. What are the most popular product keywords?

# Assuming you have a column 'product_name' in your dataset containing product names

# Extracting keywords from product names
keywords = df_cp['product_name'].str.split().explode()

# Counting the occurrences of each keyword
keyword_counts = keywords.value_counts().reset_index()

# Renaming the columns for better readability
keyword_counts.columns = ['Keyword', 'Count']

# Sorting the dataframe by count in descending order to get the most popular keywords first
keyword_counts = keyword_counts.sort_values(by='Count', ascending=False)

# Plotting the horizontal bar plot
plt.figure(figsize=(10, 8))
sns.barplot(data=keyword_counts.head(10), x='Count', y='Keyword', palette='magma')

# Adding labels and title
plt.title('Top 10 Most Popular Product Keywords')
plt.xlabel('Count')
plt.ylabel('Keyword')

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
# 7. What are the most popular product reviews?

# Convert 'rating_count' column to strings
df_cp['rating_count'] = df_cp['rating_count'].astype(str)

# Count occurrences of each review for each product
review_counts = df_cp.groupby(['product_id', 'rating_count']).size().reset_index(name='count')

# Find the most popular review for each product
popular_reviews = review_counts.loc[review_counts.groupby('product_id')['count'].idxmax()]

# Plotting the distribution of the most popular reviews
plt.figure(figsize=(10, 8))
sns.barplot(data=popular_reviews.head(10), x='count', y='product_id', hue='rating_count', dodge=False, palette='viridis')

# Adding labels and title
plt.title('Top 10 Most Popular Product Reviews')
plt.xlabel('Count')
plt.ylabel('Product ID')
plt.legend(title='Review', bbox_to_anchor=(1.05, 1), loc='upper left')

# Displaying the plot 
plt.tight_layout()
plt.show()
```

```python
# 8. What is the correlation between discounted_price and rating?

# Calculate correlation coefficient
correlation = df_cp['discounted_price_rupees'].corr(df_cp['rating'])

# Generate scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cp, x='discounted_price_rupees', y='rating')
plt.title(f'Correlation between Discounted Price and Rating: {correlation:.2f}')
plt.xlabel('Discounted Price')
plt.ylabel('Rating')
plt.grid(True)
plt.show()

# Insights:
# 1. Highest discount price is 80k and rating is 4.8
# 2. Lowest discount prive is 0 and rating is 5
# 3. Average discount prive is 10k and average rating is in between 3.8 to 4.4 
```

```python
# 9. What are the Top 5 categories based on the highest ratings?
# Group by category and calculate average rating
category_ratings = df_cp.groupby('category')['rating'].mean().sort_values(ascending=False)

# Select top 5 categories
top_5_categories = category_ratings.head(5)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_categories.values, y=top_5_categories.index, palette='viridis')
plt.title('Top 5 Categories Based on Highest Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Category')
plt.show()
```
