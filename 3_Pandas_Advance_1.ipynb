```python
# what is avg fare paid by people who survived? 

# groupby based on one column 

# group by : Interview question how group by works 

# https://www.w3resource.com/python-exercises/pandas/groupby/index.php
# group the data based on column called ID 
# if we want to group + by split by unique values of x,y,z from image given 
# next step do aggregation seperatly for x , y , z 
# combine the x,y,z 

df.groupby('Survived').mean(numeric_only=True) #for recent pandas version
```

```python
# df.groupby('Survived').mean() # may be mean was depricated
df.groupby('Survived').describe()
```
