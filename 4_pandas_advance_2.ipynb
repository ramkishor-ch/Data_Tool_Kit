```python
# Interview Questions

# 1. Left Outer Join
# 2. Right Outer Join
# 3. Inner Join
# 4. Outer Join
# 5. Cross Join
```

```python
# Merge

pd.merge(df1, df2, how = 'left') 
# All the elements from df1 and 
# common elements from df2 present in df1

# Example:
"""
Output contains all columns of df1 
    +
Common column in df2 and df1 is "key1" column, 
the common rows which are present in key1 column from df1 and df2,
remains same in key4 and key5 column of df2.

the rows of key1 column from df1,
which are not present in rows of key1 column from df2,
then rows are automatically marked as "NaN" in column "key4" and "key5" of df2.

"NaN" : empty value or None or Null Value or Not a number
"""
```

```python
df1
```

## 

```python
df2
```

```python
#merge
pd.merge(df1, df2, how = 'right')
# All the elements from df2 and 
# common elements from df1 present in df2

# Example:
"""
Output contains all columns of df2
    +
Common column in df1 and df2 is "key1" column, 
the common rows which are present in key1 column from df2 and df1,
remains same in key1 and key2 column of df1.

the rows of key1 column from df2,
which are not present in rows of key1 column from df1,
then rows are automatically marked as "NaN" in column "key1" and "key2" of df1.

"NaN" : empty value or None or Null Value or Not a number
"""
```

```python
pd.merge(df1, df2, how = 'outer') #all the keys of both the datframe
# Outer Join is combination of Left Join and Right Join
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, how = 'cross')
# Cross Join
# the first row of df1 is combined with all rows of df2
# the second row of df1 is combined with all rows of df2
# the thrird row of df1 is combined with all rows of df2
# the fourth row of df1 is combined with all rows of df2
# the fifth row of df1 is combined with all rows of df2
```

```python
df1
```

```python
df2
```

```python
# Left Join ( No Common Column in df1 and df2, how to do it?>
# Solution:

# Q. How to Join two data frames without having common column in df1 and df2?
# S: 
"""
considering key2 is common column from df1 and key4 is common column from df2

Output contains all columns of df1 
    +
Common column in df1 is "key2" column,
Common column in df2 is "key4" column,

the common rows ( which are present in key2 column from df1 and key4 column from df2 ),
remains same in key4 and key5 column of df2.

the un-common rows of key2 column from df1 and key4 column from df2,
which are not present in rows of key4 and key5 column from df2,
then rows are automatically marked as "NaN" in column "key4" and "key5" of df2.
"""

# Q. Why key1_x name is changed in df1 and key1_y name is changed in df2?
# S: One data frame cannot have same column name for two columns, so that's why merge understood
#    this common column key1 cannot merged so key1 column have to be renamed.
#    key1_x is coming from df1 and key1_y is coming from df2

# Q. Suppose if there is no common column called "key1" in df1 and df2
#    or
# Q. If I don't want to consider common column called "key1" in df1 and df2, 
#    explicitly defining common column ( key2 from df1 and key4 from df2 ) ?
pd.merge(df1, df2, how = 'left', left_on = 'key2', right_on = 'key4')

# When converted to NaN then columns in df2 are converted to float values
# If needed then convert to integer
```

```python
df1.join(df2, how = 'left', ) 
# join happens based on the indexes, column name should be not be common
# Indexs: a, b are common in df1 and df2 then remaining rows in df2 are marked as NaN
```

```python
df1.join(df2, how = 'right', )
# join happens based on the indexes, column name should be not be common
# Indexs: a, b are common in df1 and df2 then remaining rows in df1 are marked as NaN
```

```python
df1.join(df2, how = 'outer' )
# join happens based on the indexes, column name should be not be common
# Outer Join = Left Join + Right Join
```

```python
df1.join(df2, how = 'cross' )
# join happens based on the indexes, column name should be not be common
# Each Row in df1 is repeated with all rows in df2
```

```python
df1.join(df2, how = 'inner' )
# join happens based on the indexes, column name should be not be common
# Indexs: a, b are common in df1 and df2.
```
