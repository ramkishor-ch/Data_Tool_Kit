```python
# Death Event

# Indicates whether the patient died during the follow-up period (binary: 0 for no, 1 for yes).
# 1. What is the distribution of age among heart failure patients in the dataset?

import matplotlib.pyplot as plt
import seaborn as sns

# Filter the dataset to include only heart failure patients
heart_failure_patients = df_cp[df_cp['DEATH_EVENT'] == 1]  # Assuming 'DEATH_EVENT' indicates heart failure

# Plot the distribution of age using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(heart_failure_patients['age'], color='skyblue', bins=20)
plt.title('Distribution of Age Among Heart Failure Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Insights:
# 1. At the age of 60, highest heart failures occured nearly 18 patients
# 2. At the age of 42, lowest heart failures occured nearly 1 patient
# 3. Below the age of 40, No heart failures
# 4. At the age of 65, average heart failures occured nearly 9 patients
# 5. When the age is increasing after 60 the number of heart failures got reducing.
# 6. At the age of 72 to 78 consistent heart failures occures for 6 patients
# 7. At the age of 88 to 89, No heart failures

# Doubt:
# what is the blue line passing on the bars in graph if mentioned kde=True?
```

```python
# 2. How does the death rate vary with age?
# Set custom y-ticks
yticks = [0, 1]

plt.figure(figsize=(7, 2))
sns.scatterplot(x='age', y='DEATH_EVENT', data=df_cp, color='skyblue')
plt.yticks(yticks)  # Set custom y-ticks
plt.title('Death Rate Variation with Age')
plt.xlabel('Age')
plt.ylabel('Death Rate')
plt.show()

# Insights:
# 0 for no, 1 for yes
# 1. Most of the people are died in between age 57 to 60
# 2. Age increases then death rate also increases.
```

```python
# 6. How does the prevalence of high blood pressure differ between male and female patients?
# df_cp
# Calculate the total number of male and female patients
total_male_patients = df_cp[df_cp['sex'] == 1].shape[0]
total_female_patients = df_cp[df_cp['sex'] == 0].shape[0]

# Calculate the proportion of patients with high blood pressure for each gender group
male_high_bp_proportion = (df_cp[(df_cp['sex'] == 1) & (df_cp['high_blood_pressure'] == 1)].shape[0] / total_male_patients) * 100
female_high_bp_proportion = (df_cp[(df_cp['sex'] == 0) & (df_cp['high_blood_pressure'] == 1)].shape[0] / total_female_patients) * 100

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(['Male', 'Female'], [male_high_bp_proportion, female_high_bp_proportion], color=['blue', 'red'])
plt.title('Prevalence of High Blood Pressure by Gender')
plt.xlabel('sex')
plt.ylabel('Proportion of Patients with High Blood Pressure (%)')
plt.ylim(0, 100)  # Set y-axis limit to ensure percentages are properly displayed
plt.show()


# Insights:
# 1. Compare to male, female is having high blood pressure
# 2. 40 females having high blood pressure
# 3. 32 males having high blood pressure
```

```python
# 7. what is the relationship between smoking habits and the occurrencs of heart failure?

# Filter the data to include only heart failure cases
heart_failure_cases = df_cp[df_cp['DEATH_EVENT'] == 1]

# Count the occurrences of heart failure among smokers and non-smokers
heart_failure_counts = heart_failure_cases['smoking'].value_counts()

# Create a bar plot
plt.figure(figsize=(8, 6))
heart_failure_counts.plot(kind='bar', color=['blue', 'green'])
plt.title('Occurrences of Heart Failure Among Smokers and Non-Smokers')
plt.xlabel('Smoking Habit (0: Smoker, 1: Non-Smoker)')
plt.ylabel('Number of Heart Failure Cases')
plt.xticks([0, 1], ['Smoker', 'Non-Smoker'], rotation=0)
plt.show()

# Insights:
# 1. Highest 60+ heart failure occured among smokers.
# 2. Double times of heart failures occured among smokers than non smokers
# 3. Compared to smokers, only 30 heart failures occured among non smokers
```

```python
# 8. Are there any noticeable patterns in the distribution of death events across different age groups?

# Create a box plot to visualize the distribution of death events across different age groups
plt.figure(figsize=(10, 6))
sns.boxplot(x='DEATH_EVENT', y='age', data=df_cp)
plt.title('Distribution of Death Events Across Age Groups')
plt.xlabel('Death Event (0: No, 1: Yes)')
plt.ylabel('Age')
plt.show()

# Insights:
# 1. There seems to be a trend towards higher death events as age increases, particularly among older age groups.
# 2. There are some outliers in the older age groups, suggesting that while death events are more common among older individuals, 
#    there are exceptions where death events occur at younger ages as well.
```

```python
# 9. Is there any significant difference in ejection fraction between patients with and without diabetes?

# Create a box plot to compare ejection fraction between patients with and without diabetes
plt.figure(figsize=(10, 6))
sns.boxplot(x='diabetes', y='ejection_fraction', data=df_cp)
plt.title('Comparison of Ejection Fraction Between Patients with and without Diabetes')
plt.xlabel('Diabetes (0: No, 1: Yes)')
plt.ylabel('Ejection Fraction')
plt.show()

# Insights:
# 1. When ejection fraction is 80 there are no diabetes patients.
# 2. When ejection fraction is below 65 there are diabetes patients.
```
